```python
import numpy as np
K=3
X=pd.read_csv('../data/visible_clusters.csv').to_numpy()
# mu_0=[np.mean(X)-np.std(X),np.mean(X),np.mean(X)+np.std(X)]
mu_new = np.random.randint(0, 10, size=(3, 2))
# mu_new=-np.ones((3,2))*5

mu_old=np.ones((3,2))*np.inf

count=0
while np.linalg.norm(mu_old-mu_new)>0.001 and count<100:
    count+=1
    mu_old=mu_new.copy()
    
    # print(count,f' mu_old is {mu_old}')
    R=np.zeros((K,X.shape[0]))
    for i in range((X.shape[0])):
        index=np.argmin([np.linalg.norm(X[i]-mu_old[j]) for j in range(K)])
        R[index,i]=1
    index_array=np.argmax(R, axis=0)
    num=np.matmul(R,X)
    den=R.sum(axis=1).reshape(-1,1)
    plot_data_2d(X,mu_new,index_array)
    print(f' mu_new is {mu_new}')
    mu_new= np.divide(num,den,out=np.zeros_like(num, dtype=float),where=(den != 0) & (num != 0))
    
        
```
